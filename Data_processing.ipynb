```python
import pandas as pd

data = pd.DataFrame({
    'col1': [1,2,3,4],
    'col2': [5,6,7,8],
    'col3': [9,10,11,12]
})
```

```python
# 選擇第一列資料
row = data.loc[0]
print(row)
```

```python
import pandas as pd
# 選擇第一列和第二列資料
data = pd.DataFrame({
    'col1': [1,2,3,4],
    'col2': [5,6,7,8],
    'col3': [9,10,11,12]
})

rows = data.loc[[0,1]]
print(rows)
# 選擇第一列和第二列 以及第一欄和第二欄位資料

data.sort_values(['col1', 'col2'], ascending=[True, False])
```

```python
import pandas as pd
# Pandas provide "query" 用於資料過濾
data = pd.DataFrame({
    'col1': [1,12,3,14],
    'col2': [5,6,7,8],
    'col3': [9,10,11,12]
})
subset = data.query('col1 > 10')
# 整個欄位中只選取 col1 > 10 的數列
print(subset)
# 整個欄位中只選取 col1 > 10 的數列 跟在 col2 是在選取20以下
subset = data[(data['col1'] < 10) & (data['col2'] < 20)]
```
